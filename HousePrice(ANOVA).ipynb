```python
missing_value_percentage_df=pd.DataFrame({'Column_Name':house_price_df.columns,
                                          'Missing_value_count':house_price_df.isnull().sum(),
                                          'Missing_value_percentage':house_price_df.isnull().sum()/house_price_df.shape[0]*100})


missing_value_percentage_df=missing_value_percentage_df.sort_values(by='Missing_value_percentage',ascending=False)
```

```python
cols_to_drop=[]

cols_with_na_grt_50=list(missing_value_percentage_df[missing_value_percentage_df['Missing_value_percentage']>50].index)

for col in house_price_df.columns:
    if house_price_df[col].nunique()==house_price_df.shape[0] or house_price_df[col].nunique()==1:
        cols_to_drop.append(col)
        
        
columns_to_drop=cols_to_drop+cols_with_na_grt_50
```

```python
house_price_df.drop(columns=columns_to_drop,inplace=True)
```

```python
cat_cols=[col for col in house_price_df.columns if house_price_df[col].dtype == 'object']
num_cols=[col for col in house_price_df.columns if house_price_df[col].dtype == 'int64' or 
                                                    house_price_df[col].dtype == 'float64']
```

```python
#house_price_df[cat_cols]

#ANOVA
import scipy.stats as stats

def anova(col):
    categories=house_price_df[col].unique()
    return stats.f_oneway(*(house_price_df[house_price_df[col] == category]['SalePrice'] for category in categories))[1]


```

```python
anova_df=pd.DataFrame({'Column_Name':cat_cols,'p_value':[anova(col) for col in cat_cols]})
```
